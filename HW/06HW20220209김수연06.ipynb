```python
# 1부터 10,000까지 8이라는 숫자가 총 몇 번 나오는가?
# 8이 포함되어 있는 숫자의 갯수를 카운팅 하는 것이 아니라 8이라는 숫자를 모두 카운팅 해야
# 한다.
# (※ 예를들어 8808은 3, 8888은 4로 카운팅 해야 함, 구글 입사문제)
# 1. 파이썬 코드와 결과 포함한 사진 제출
# 2. 순서도나 수도코드까지 작성하면 extra credit
# 3. 각 행마다 주석 달기 필수, 미 기재 시 감점

num1=0
num2=0
num3=0
num4=0
#자리수별로 8의 개수를 따로 센다.
for i in range(1,10001): #1부터 10000까지 반복한다.
    if (i > 0) and (i < 10): # i가 한 자리수 숫자라면
        x = str(i) #i를 string으로 바꾸고
        if '8' in x[-1]: #1의자리가 8이라면
            num1 = num1 + 1 #1의 자리 개수에 1을 추가한다.
        
    if (i > 10) and (i < 100) : # i가 두 자리수 숫자라면
        x = str(i) #i를 string으로 바꾸고
        if '8' in x[-1]: #1의자리가 8이라면
            num1 = num1 + 1 #1의 자리 개수에 1을 추가한다.
        if '8' in x[-2]: #10의 자리가 8이라면
            num2 = num2 + 1 #10의 자리 개수에 1을 추가한다.

    if (i > 100) and (i < 1000) : # i가 세 자리수 숫자라면
        x = str(i) #i를 string으로 바꾸고
        if '8' in x[-1]: #1의자리가 8이라면
            num1 = num1 + 1 #1의 자리 개수에 1을 추가한다.
        if '8' in x[-2]: #10의 자리가 8이라면
            num2 = num2 + 1 #10의 자리 개수에 1을 추가한다.
        if '8' in x[-3]: #1의자리가 8이라면
            num3 = num3 + 1 #100의 자리 개수에 1을 추가한다.

    if (i > 1000) and (i < 10000) : # i가 네 자리수 숫자라면
        x = str(i) #i를 string으로 바꾸고
        if '8' in x[-1]: #1의 자리가 8이라면
            num1 = num1 + 1 #1의 자리 개수에 1을 추가한다.
        if '8' in x[-2]:#10의 자리가 8이라면 
            num2 = num2 + 1 #10의 자리 개수에 1을 추가한다.
        if '8' in x[-3]: #100의 자리가 8이라면
            num3 = num3 + 1 #100의 자리 개수에 1을 추가한다.
        if '8' in x[-4]: #1000의 자리가 8이라면
            num4 = num4 + 1 #1000의 자리 개수에 1을 추가한다.
    
print(num1+num2+num3+num4) # 1, 10 ,100, 1000의 자리의 8 개수를 모두 더한다.
```
