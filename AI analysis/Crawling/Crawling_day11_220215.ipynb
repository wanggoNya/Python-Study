```python
html = requests.get('https://music.bugs.co.kr/chart')
print(html.text)
```

```python
soup = bs(html.text) # 글자만 가져오는 위 태그와 달리 이번 명령어는 html의 진짜 기능까지 살림
# print(soup) 
```

```python
len(soup.select('table.byChart > tbody > tr')) # 태그 붙은 부분들은 모두 추출하는데 범위를 축소했음
```

```python

songs = soup.select('table.byChart > tbody > tr') #버그 줄여줌 #최종 100개 추출 완료
songs
```

```python
song_data = []
rank = 0

songs = soup.select('table.byChart > tbody > tr') #버그 줄여줌 #최종 100개 추출 완료

for song in songs:
    singer = song.select('p.artist > a')[0].text # 계속 범위 줄여서 태그를 뽑아냄, [0].text 를 이용해서 가수 이름 뽑아냄
    title = song.select('p.title > a')[0].text
    rank += 1
    song_data.append([rank, title, singer])
    
print(song_data)
    

```

```python
import pandas as pd


df = pd.DataFrame(song_data, columns = ['순위', '타이틀', '가수'])
df
```

```python
df.to_excel("bugs.xlsx", index = False)
```
